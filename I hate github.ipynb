```python
y1,sr1 = librosa.load('synthesizef.wav')
y2,sr2 = librosa.load('synthesizem.wav')
```

```python
y2
```

```python
if(len(y1)<len(y2)):
    y1=np.concatenate([y1, np.zeros(len(y2)-len(y1))])
else:
     y2=np.concatenate([y2, np.zeros(len(y1)-len(y2))])
```

```python
#obviously don't do this but I just wanted to put something down lol
reg = LinearRegression()
reg.fit(y1.reshape(-1, 1),y2.reshape(-1, 1))
```

```python
result = reg.predict(y1.reshape(-1, 1))
```

```python
sf.write('file_trim_5s.wav', result, sr2)
```
